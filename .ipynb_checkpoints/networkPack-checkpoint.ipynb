```python
import struct
import binascii
def float_to_hex(f):
    return binascii.hexlify(struct.pack('<f', f))
float_to_hex(17.5) # 0x418c0000
print(float_to_hex(17.5).decode("utf-8"))
```

```python
#Reverse order of N (2-bit by 2-bit)
```

```python
def reverse_N(N):
    print(N)
    dataLine_hex = str(hex(N))[2:]
    dataLine_hex = dataLine_hex.rjust(8, '0')
    print(dataLine_hex)
    dataLine_hex_reverse = ""
    for i in range(4):
        dataLine_hex_reverse += dataLine_hex [6-2*i: 8-2*i]
    return dataLine_hex_reverse
```

```python
#10 batches in one packet, each has 799 data (50 dataLines)
```

```python
timeInfo = ["0020", "0000"]
packet_1 = timeInfo 
dataLine_perBatch = 50
for totalBatch in range(10):
    for lineNum in range(dataLine_perBatch):
        inst = ""
        if (lineNum == dataLine_perBatch - 1 and totalBatch == 9):
            inst = inst + '1'
        else:
            inst = inst + '0'
        for floatContent in range(16):
            if(lineNum == 0 and floatContent ==15):
                inst = inst + reverse_N(16*dataLine_perBatch - 1)
            else:
                inst = inst + float_to_hex(0.02*totalBatch +0.5*lineNum+ 0.3*floatContent).decode("utf-8")
        if lineNum==0:
            print( "lineNum: ", lineNum, "  -------- batchNum: ", totalBatch, "\n", inst)
        packet_1 = packet_1 + [inst]
        #print(len(inst))
print(len(packet_1))
```

```python
# Result Verification packet_1
for i in range(10):
    total = 0
    for j in range(dataLine_perBatch):
        for k in range(16):
            total += 0.02* i + 0.5*k +0.5*j
    print(total/(16*dataLine_perBatch-1), "----float32: ", float_to_hex(total/(16*dataLine_perBatch-1)))
```

```python
#10 batches in two packets, each has 479 data (30 dataLines), 300 dataLines in total
#packet_2 has 200 dataLines (5 batches + 20 dataLines), packet_3 has 100 dataLines(3 batches and 10 datalines)
```

```python
timeInfo = ["0020", "0000"]
packet_2_3 = timeInfo
dataLine_perBatch = 30
for totalBatch in range(10):
    for lineNum in range(dataLine_perBatch):
        inst = ""
        if(lineNum == 20 and totalBatch == 6): #packet_3 starts
            packet_2_3 += timeInfo
            
        if ((lineNum == 19 and totalBatch == 6) or (lineNum == 29 and totalBatch == 9)): #packet_2 or packet_3 end
            inst = inst + '1'   
        else:
            inst = inst + '0'
        for floatContent in range(16):
            if(lineNum == 0 and floatContent ==15):
                inst = inst + reverse_N(16*dataLine_perBatch - 1)
            else:
                inst = inst + float_to_hex(totalBatch +0.5*lineNum+ 0.3*floatContent).decode("utf-8")
        if lineNum==0:
            print( "lineNum: ", lineNum, "  -------- batchNum: ", totalBatch, "\n", inst)
        packet_2_3 = packet_2_3 + [inst]
print(len(packet_2_3))
```

```python
# Result Verification packet_2_3
for i in range(10):
    total = 0
    for j in range(dataLine_perBatch):
        for k in range(16):
            total += i + 0.5*k +0.3*j
    print(total/(16*dataLine_perBatch-1), "----float32: ", float_to_hex(total/(16*dataLine_perBatch-1)))
```

```python
import os
if os.path.exists("session-data-in.txt"):
    os.remove("session-data-in.txt")
networkPacket = open("session-data-in.txt", "w")
for lineContent in packet_1 + packet_2_3:
    networkPacket.writelines(lineContent)
    networkPacket.write("\n")
networkPacket.close()
```
